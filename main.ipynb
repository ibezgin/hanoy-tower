```python
def hanoi(n, source, target, auxiliary):
    if n > 0:
        # Перемещение n-1 дисков с начального стержня на вспомогательный стержень
        hanoi(n-1, source, auxiliary, target)
        
        # Перемещение n-го диска с начального стержня на целевой стержень
        print(f"Перемещаем диск {n} со стержня {source} на стержень {target}")
        
        # Перемещение n-1 дисков с вспомогательного стержня на целевой стержень
        hanoi(n-1, auxiliary, target, source)
        
# Запрашиваем количество дисков у пользователя
num_disks = int(input("Введите количество дисков на стержнях: "))

# Вызываем функцию hanoi для решения головоломки
hanoi(num_disks, 'A', 'C', 'B')
```
